```python
from wordcloud import WordCloud , STOPWORDS
STOPWORDS.update(['feel','feeling','im',",","t","u","2","'","&amp;","-","...","s"])
def extract_most_used_word(text, stopwords=STOPWORDS, words=1):
    """Return the wanted number of most used word of a given text and their number of occurences"""
    words_list=[]
    # iterate through the csv file
    for val in text:
        # typecaste each val to string
        val = str(val)
        # split the value
        tokens = val.split()
        # Converts each token into lowercase
        for i in range(len(tokens)):
            tokens[i] = tokens[i].lower()
            if not tokens[i] in stopwords:
                words_list.append(tokens[i])
    occurence_count = Counter(words_list)
    return np.array(occurence_count.most_common())[:words]
```

```python
from collections import Counter
for emotion in df_dataK["Emotion"].unique():
    print(emotion)
    emotion_words = extract_most_used_word(df_dataK[df_dataK["Emotion"]== emotion].Text,words=20)
    make_plot( emotion_words[:,0] , emotion_words[:,1].astype("int"), barplot ,"Kaggle"+emotion, xlabel="Emotion", ylabel="Quantité", override=True) 
```
